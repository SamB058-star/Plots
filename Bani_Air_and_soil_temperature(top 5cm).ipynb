```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates


files_path = 'D:\\Stations_data\\Bani\\all_months_copy'
os.chdir(files_path)
df = pd.read_csv('till_january.csv', sep=',')
df['date'] = pd.to_datetime(df.date, format = "%d.%m.%Y %H:%M:%S")
df = df.set_index('date')
df = df.sort_index()
```

```python
df_1hr = df.resample('60min').mean()
df2 = pd.read_csv('Temp_Rain_Bani.csv', sep = ',')
df2['Date & Time'] = pd.to_datetime(df2['Date & Time'], format = "%d/%m/%Y - %H:%M:%S")
df2 = df2.set_index('Date & Time')
df2_1hr = df2.resample('60min').mean()
df3 = pd.read_csv('Rainfall_data_new.csv', sep = ',', usecols = ['Date & Time', 'Int_Rainfall (mm)'])
df3['Date & Time'] = pd.to_datetime(df3['Date & Time'])
df3 = df3.set_index('Date & Time')
rainfall_1hr = df3.resample('60min').mean()
rainfall_1hr.fillna(0.0, inplace = True)
#rainfall_1hr
```

```python
%matplotlib inline
#plot for sensor 1
fig, ax1 = plt.subplots(figsize=(12, 5))
#ax1b = ax1.twinx()
ax1.plot(df_1hr.index, df_1hr.S1Temp, '-', color = 'red', label = 'Soil Temperature 1hr')
ax1.plot(df2_1hr.index, df2_1hr['Temp (0C)'], '--', color = 'grey', label = 'Air Temperature 1hr')
#ax1.set_title('Soil Temperature and Air Temorature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature [$^\circ$C]')
ax1.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
ax1.xaxis.set_major_formatter(DateFormatter("%d-%m-%Y"))
ax1.tick_params(axis='x', labelrotation = 45)
ax1.legend(loc='upper right', shadow=True, ncol=2)
plt.tight_layout()
plt.title("Sensor-11 Soil Temperature and Air Temorature", fontweight="bold")
plt.show()
```
